```python
data = pd.read_csv('/Users/maosa/fitr/data.csv', sep=',', )
```

```python
workouts = data['Workout'].unique()

workouts
```

```python
chest = [
    'Chest', 'Insane Chest and Dips', 'Insane Chest and Biceps', 'Insane Chest',
    'Chest and Biceps', 'Chest and Triceps and Core', 'Chest and Core', 'Chest and Core and Rotator Cuffs',
    'Half Squats and Chest and Core', 'Insane Chest and Push and Core', 'Insane Chest and Push',
    'Multisport (Chest)'
]

back = [
    'Back', 'Pull 1', 'Pull 2', 'Cobra Back', 'Back and Biceps and Core', 'Back and Biceps',
    'Back and Knee Conditioning', 'Cobra Back and Core', 'Cobra Back and Pull', 'Back and Core',
    'Multisport (Back)'
]

legs = ['Legs', 'Legs and Core', "Can't Walk and Core", "Can't Walk", 'Multisport (Legs)']

compound = [
    'Compound', 'Texas', 'Chest and Back and Knee Conditioning',
    'Compound and Core', 'Texas and Dips and Core',
    'Chest and Back', 'Compound 1', 'Compound 2', 'Compound 3', 'Cross training', 'Multisport', 'Bouldering'
]

cardio = ['Running', 'Cardio', 'Basketball', 'Surfing']

other = [
    'Shoulders and Arms', 'Dorians Challenge', 'Upper Body'
]

rest = 'Rest'

double = [s for s in workouts if '+' in s]

if (len(workouts) != len(chest) + len(back) + len(legs) +
    len(compound) + len(cardio) + len(rest) + len(double)):
    
    miss = len(workouts) - (len(chest) + len(back) + len(legs) +
    len(compound) + len(cardio) + len(rest) + len(double))
    
    print('Mismatch of', miss)
```

```python
# Simplify workout names

simple = []

for w in range(len(data['Workout'])):
    
    workout = data['Workout'][w]
    
    if workout in chest:
        simple.append('Chest')
    elif workout in back:
        simple.append('Back')
    elif workout in legs:
        simple.append('Legs')
    elif workout in compound:
        simple.append('Compound')
    elif workout in cardio:
        simple.append(workout)
    elif workout in rest:
        simple.append('Rest')
    elif '+' in workout:
        simple.append(workout)
    else:
        simple.append('Other')
        
if len(simple) == len(data):
    data['simple'] = simple

# Categorise workoute

category = []

for w in range(len(data['Workout'])):
    
    workout = data['Workout'][w]
    
    if workout in chest:
        category.append('chest')
    elif workout in back:
        category.append('back')
    elif workout in legs:
        category.append('legs')
    elif workout in compound:
        category.append('compound')
    elif workout in cardio:
        category.append('cardio')
    elif workout in rest:
        category.append('rest')
    elif '+' in workout:
        category.append('double_workout')
    else:
        category.append('other')
        
if len(category) == len(data):
    data['Category'] = category

# Simplify workout names

data['Workout'] = data['simple']

data.drop(labels='simple', axis=1, inplace=True)

# Convert dates to datetime
    
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')

# Check dates and reverse if necessary

first = data['Date'][0]
last = data['Date'][data.shape[0] - 1]

if first < last:
    data.sort_index(ascending=False, inplace=True)
else:
    pass

# Make colnames lowercase

data.columns = [x.lower() for x in list(data.columns)]
```
